```python
import folium
import pandas as pd
import json

# 지도 데이터 파일 경로
geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

# 기온 데이터 프레임
df_temperature = df.최고기온

# 열 이름 수정
df_temperature.columns = ['구별', '평균온도']

# 지도 생성
m = folium.Map(location=[37.566345, 126.9778931], tiles='Stamen Terrain')

# Choropleth 레이어 추가
folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df_temperature,
    columns=['구별', '평균온도'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='평균온도(℃)'
).add_to(m)

# 지도 출력
m
```

```python

import matplotlib.pyplot as plt
plt.rc('font',family ='NanumBarunGothic')

#온도랑 인구밀도
import matplotlib.pyplot as plt

# 데이터
a = population.sort_index()
x = a.index[0:25]
y1 = a['인구밀도 ']
b = df.sort_index()
y2 = b['최고기온']

# 그래프 그리기
fig, ax1 = plt.subplots()  # 첫 번째 y축을 사용하는 막대 그래프를 그릴 축
ax2 = ax1.twinx()  # 두 번째 y축을 사용하는 꺽은선 그래프를 그릴 축

# 막대 그래프
ax1.bar(x, y1, color='b', alpha=0.5, label='Bar Graph')
ax1.set_ylabel('population')

# 꺽은선 그래프
ax2.plot(x, y2, color='g', marker='o', linestyle='-', label='Line Graph')
ax2.set_ylabel('temp')
ax1.set_xticks(x)
ax1.set_xticklabels(a.index[0:25], rotation='vertical')


# 범례
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# 축 레이블
ax1.set_xlabel('district')

# 그래프 표시
plt.show()

# 온도랑 면적
import matplotlib.pyplot as plt

# 데이터
a = park.sort_index()
x = a.index[0:25]
y1 = a['면적']
b = df.sort_index()
y2 = b['최고기온']

# 그래프 그리기
fig, ax1 = plt.subplots()  # 첫 번째 y축을 사용하는 막대 그래프를 그릴 축
ax2 = ax1.twinx()  # 두 번째 y축을 사용하는 꺽은선 그래프를 그릴 축

# 막대 그래프
ax1.bar(x, y1, color='b', alpha=0.5, label='Bar Graph')
ax1.set_ylabel('population')

# 꺽은선 그래프
ax2.plot(x, y2, color='g', marker='o', linestyle='-', label='Line Graph')
ax2.set_ylabel('temp')

ax1.set_xticks(x)
ax1.set_xticklabels(a.index[0:25], rotation='vertical')

# 범례
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# 축 레이블
ax1.set_xlabel('district')

# 그래프 표시
plt.show()
```
